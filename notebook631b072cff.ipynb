```python



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



sns.set(style="whitegrid", palette="muted", font_scale=1.2)


df = pd.read_csv("/kaggle/input/air-quality-environmental-data-tabular-2025/AirQuality_data.csv")



print("Shape of dataset:", df.shape)   # عدد الصفوف والأعمدة
print("\nFirst 5 rows:\n", df.head())  # أول 5 صفوف
print("\nInfo:\n")
df.info()                              # معلومات الأعمدة وأنواعها
print("\nDescriptive Statistics:\n", df.describe())  # إحصائيات وصفية


print("\nMissing values per column:\n", df.isnull().sum())


print("\nClass distribution:\n", df['class'].value_counts())

plt.figure(figsize=(6,6))
df['class'].value_counts().plot(kind='bar', color=['skyblue','salmon'])
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()




plt.figure(figsize=(8,5))
sns.histplot(df['TEMP'], bins=20, kde=True)
plt.title("Distribution of Temperature")
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(df['Hum'], bins=20, kde=True)
plt.title("Distribution of Humidity")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=df['CO2'])
plt.title("Boxplot of CO2")
plt.show()


plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="TEMP", y="CO2", hue="class", alpha=0.7)
plt.title("CO2 vs TEMP by Class")
plt.show()


plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
